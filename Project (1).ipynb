# ïƒ˜	Is there a particular ad topic line that receives more clicks than others?

```python
# Group the data by ad_topic_line and calculate the mean click-through rate (CTR)
ctr_by_topic = ad.groupby('Ad Topic Line')['Clicked on Ad'].mean().reset_index()
ctr_by_topic
```

```python
# Sort the results in descending order to find the most-clicked topics
sorted_ctr_by_topic =ctr_by_topic.sort_values(by='Clicked on Ad', ascending=False)
sorted_ctr_by_topic
```

```python
# Print the topics with the highest CTR
print("Ad Topic Lines with the Highest Click-Through Rates:")
print(sorted_ctr_by_topic.head())
```

```python
head_of_sorted_topic  = sorted_ctr_by_topic.head()
head_of_sorted_topic
```

```python
# Plot the bar diagram
plt.figure(figsize=(12, 6))
plt.bar(head_of_sorted_topic['Ad Topic Line'], head_of_sorted_topic['Clicked on Ad'],color=["r","b","black","orange","purple"])
plt.xticks(rotation=90)
plt.xlabel('Ad Topic Line',size=15)
plt.ylabel('Click-Through Rate (CTR)',size = 15)
plt.title('Click-Through Rate by Ad Topic Line',size = 20)
plt.show()
```

# Is there a relationship between daily time spent on the site and Daily Internet usage those clicking on the ad?

```python
clicked_ad = ad[ad["Clicked on Ad"]==1]
clicked_ad
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(clicked_ad["Daily Time Spent on Site"],clicked_ad["Daily Internet Usage"],label="Clicked on Ad")
plt.xlabel('Daily Time Spent on Site (minutes)')
plt.ylabel('Clicked on Ad')
plt.title('Relationship Between Daily Time Spent and Clicking on Ad')
plt.legend()
plt.show()
```

Interpretation:
In the scatter plot, we can observe a generally positive correlation between daily time spent on the site and daily internet usage for those clicking on the ad. As daily time spent on the site increases, there is a tendency for daily internet usage to also increase. This suggests that users who spend more time on the site tend to use more internet data when clicking on ads.

However, it's important to note that the relationship is not perfectly linear, as there is some spread of data points. Other factors may also influence daily internet usage, and this scatter plot provides an initial visual exploration of the relationship. Further statistical analysis, such as calculating correlation coefficients, could provide a more precise assessment of the strength and direction of the relationship.

# Using a Pie chart to Predict the percentage of Clicked on Ad

```python
clicked_ad_counts = ad["Clicked on Ad"].value_counts()
clicked_ad_counts
```

```python
labels = ["clicked Ad","Not clicked Ad"]
plt.pie(clicked_ad_counts,labels=clicked_ad_counts.index,autopct="%1.1f%%")
plt.legend(labels,loc = "best")
plt.figure(figsize=(18,8))
```

In our Ad campaign analysis, we created a pie chart to visualize the distribution of clicks on the ad. The chart reveals the following insights:

Clicked on Ad: The pie chart illustrates the proportion of users who clicked on the ad compared to those who did not. It shows that (50.1%) of users clicked on the ad, indicating the effectiveness of our campaign in capturing user engagement.

Not Clicked on Ad: The remaining percentage (100% - 50.1%) represents users who did not click on the ad.
